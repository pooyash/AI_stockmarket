```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import shap

# Step 1: Load Bitcoin Data (yfinance)
bitcoin_data = yf.download('BTC-USD', start='2022-09-07', end='2024-09-15')
bitcoin_data = bitcoin_data[['Close', 'Volume']].copy()

# Step 2: Calculate Behavioral Metrics
bitcoin_data['SMA_20'] = bitcoin_data['Close'].rolling(window=20).mean()
bitcoin_data['EMA_20'] = bitcoin_data['Close'].ewm(span=20, adjust=False).mean()
bitcoin_data['Volatility'] = bitcoin_data['Close'].rolling(window=20).std()  # Volatility as a behavioral metric

# Bollinger Bands
bitcoin_data['Bollinger_Upper'] = bitcoin_data['SMA_20'] + (bitcoin_data['Volatility'] * 2)
bitcoin_data['Bollinger_Lower'] = bitcoin_data['SMA_20'] - (bitcoin_data['Volatility'] * 2)

# Drop NaN values from rolling calculations
bitcoin_data.dropna(inplace=True)

# Step 3: Normalize Data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(bitcoin_data[['Close', 'Volume', 'SMA_20', 'EMA_20', 'Volatility', 'Bollinger_Upper', 'Bollinger_Lower']])

# Step 4: Autoencoder for Anomaly Detection
input_dim = scaled_data.shape[1]
autoencoder = Sequential([
    Dense(64, activation='relu', input_shape=(input_dim,)),
    Dense(32, activation='relu'),
    Dense(64, activation='relu'),
    Dense(input_dim, activation='sigmoid')
])

autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(scaled_data, scaled_data, epochs=20, batch_size=32)

# Calculate reconstruction error for anomaly detection
reconstructed = autoencoder.predict(scaled_data)
reconstruction_error = np.mean(np.abs(scaled_data - reconstructed), axis=1)

# Set anomaly threshold (95th percentile of reconstruction error)
anomaly_threshold = np.percentile(reconstruction_error, 95)
anomalies = reconstruction_error > anomaly_threshold
bitcoin_data['Anomaly'] = anomalies

# Step 5: Visualize Behavioral Metrics and Anomalies
plt.figure(figsize=(12, 7))
plt.plot(bitcoin_data.index, bitcoin_data['Close'], label='Real Price', color='blue')
plt.plot(bitcoin_data.index, bitcoin_data['SMA_20'], '--', label='Herd Behavior (SMA)', color='orange')
plt.plot(bitcoin_data.index, bitcoin_data['EMA_20'], '--', label='Momentum (EMA)', color='purple')
plt.fill_between(bitcoin_data.index, bitcoin_data['Bollinger_Upper'],
                 bitcoin_data['Bollinger_Lower'], color='gray', alpha=0.2, label='Fear and Greed (Bollinger Bands)')
plt.scatter(bitcoin_data.index[anomalies], bitcoin_data['Close'][anomalies], color='red', label='Anomalies', s=100)
plt.title('Anomalies with Behavioral Metrics')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

# Step 6: Visualize Each Anomaly Context Separately
for idx in np.where(anomalies)[0]:
    plt.figure(figsize=(10, 5))
    start_idx = max(0, idx - 10)
    end_idx = min(len(bitcoin_data) - 1, idx + 10)
    anomaly_range = bitcoin_data.index[start_idx:end_idx + 1]

    # Plot the local context of the anomaly
    plt.plot(anomaly_range, bitcoin_data['Close'][start_idx:end_idx + 1], label='Actual Price', color='blue')
    plt.scatter(bitcoin_data.index[idx], bitcoin_data['Close'].iloc[idx], color='red', label='Anomaly', s=100)
    plt.plot(anomaly_range, bitcoin_data['SMA_20'][start_idx:end_idx + 1], '--', label='Herd Behavior (SMA)', color='orange')
    plt.plot(anomaly_range, bitcoin_data['EMA_20'][start_idx:end_idx + 1], '--', label='Momentum (EMA)', color='purple')
    plt.fill_between(anomaly_range, bitcoin_data['Bollinger_Upper'][start_idx:end_idx + 1],
                     bitcoin_data['Bollinger_Lower'][start_idx:end_idx + 1], color='gray', alpha=0.2, label='Fear and Greed (Bollinger Bands)')
    plt.title(f'Anomaly Context - Date: {bitcoin_data.index[idx]}')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.show()

# Step 7: SHAP for Explainable AI
explainer = shap.KernelExplainer(autoencoder.predict, scaled_data)
shap_values = explainer.shap_values(scaled_data)

# SHAP summary plot for explainability with behavioral names
shap.initjs()
shap.summary_plot(shap_values, pd.DataFrame(scaled_data, columns=['Close', 'Volume', 'SMA_20', 'EMA_20', 'Volatility', 'Bollinger_Upper', 'Bollinger_Lower']))
```
