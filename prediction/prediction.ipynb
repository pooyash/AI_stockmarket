MACD

```python


# Read the CSV file
df = pd.read_csv('/content/S_Mobarakeh Steel.csv')

# Clean up column names
new_column_names = ['TICKER', 'Date', 'FIRST', 'HIGH', 'LOW', 'CLOSE', 'VALUE', 'VOL', 'OPENINT', 'PER', 'OPEN', 'LAST']
df.columns = new_column_names

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')

# Calculate the MACD
# Calculate the short-term (12-day) EMA
df['EMA12'] = df['CLOSE'].ewm(span=12, adjust=False).mean()
# Calculate the long-term (26-day) EMA
df['EMA26'] = df['CLOSE'].ewm(span=26, adjust=False).mean()
# Calculate the MACD line
df['MACD'] = df['EMA12'] - df['EMA26']
# Calculate the signal line (9-day EMA of the MACD line)
df['Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

# Plotting the MACD
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['MACD'], label='MACD', color='blue')
plt.plot(df['Date'], df['Signal'], label='Signal Line', color='orange')
plt.title('MACD and Signal Line', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('MACD', fontsize=14)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)  # Add a horizontal line at 0
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python


# Read the CSV file
df = pd.read_csv('/content/S_Mobarakeh Steel.csv')

# Clean up column names
new_column_names = ['TICKER', 'Date', 'FIRST', 'HIGH', 'LOW', 'CLOSE', 'VALUE', 'VOL', 'OPENINT', 'PER', 'OPEN', 'LAST']
df.columns = new_column_names

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')

# Set the number of periods for the Stochastic Oscillator
n = 14  # For %K
d = 3   # For %D

# Calculate the lowest low and highest high over the last n periods
df['Lowest_Low'] = df['LOW'].rolling(window=n).min()
df['Highest_High'] = df['HIGH'].rolling(window=n).max()

# Calculate %K
df['%K'] = (df['CLOSE'] - df['Lowest_Low']) / (df['Highest_High'] - df['Lowest_Low']) * 100

# Calculate %D as a moving average of %K
df['%D'] = df['%K'].rolling(window=d).mean()

# Plotting the Stochastic Oscillator
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['%K'], label='%K', color='blue')
plt.plot(df['Date'], df['%D'], label='%D', color='orange')
plt.title('Stochastic Oscillator', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Stochastic Oscillator (%)', fontsize=14)
plt.axhline(80, color='gray', linestyle='--', linewidth=0.5)  # Overbought threshold
plt.axhline(20, color='gray', linestyle='--', linewidth=0.5)  # Oversold threshold
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python


# Read the CSV file
df = pd.read_csv('/content/S_Mobarakeh Steel.csv')

# Clean up column names
new_column_names = ['TICKER', 'Date', 'FIRST', 'HIGH', 'LOW', 'CLOSE', 'VALUE', 'VOL', 'OPENINT', 'PER', 'OPEN', 'LAST']
df.columns = new_column_names

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')

# Set the period for RSI
n = 14

# Calculate price changes
df['Price Change'] = df['CLOSE'].diff()

# Calculate gains and losses
df['Gain'] = df['Price Change'].where(df['Price Change'] > 0, 0)
df['Loss'] = -df['Price Change'].where(df['Price Change'] < 0, 0)

# Calculate the average gain and average loss
avg_gain = df['Gain'].rolling(window=n, min_periods=1).mean()
avg_loss = df['Loss'].rolling(window=n, min_periods=1).mean()

# Calculate RS (Relative Strength)
rs = avg_gain / avg_loss

# Calculate RSI
df['RSI'] = 100 - (100 / (1 + rs))

# Plotting the RSI
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['RSI'], label='RSI', color='blue')
plt.axhline(70, color='red', linestyle='--', linewidth=0.5)
plt.axhline(30, color='green', linestyle='--', linewidth=0.5)
plt.title('Relative Strength Index (RSI)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('RSI', fontsize=14)
plt.ylim(-10, 110)
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Define sequence length and number of features
sequence_length = 10
num_features = len(df1.columns)  # Number of preprocessed features

# Create sequences
sequences = []
target = []

for i in range(len(df1) - sequence_length):
    seq = df1.iloc[i:i+sequence_length].values
    target_value = df1.iloc[i+sequence_length]['CLOSE']
    sequences.append(seq)
    target.append(target_value)

sequences = np.array(sequences)
target = np.array(target)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(sequences, target, test_size=0.2, shuffle=False)

```

آموزش مدل LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Build LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(sequence_length, num_features)))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)


```

```python
from sklearn.metrics import mean_squared_error

#  predictions on the test data
y_pred = model.predict(X_test)

#  Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

#  Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
```

```python


from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
import numpy as np

# Create and train the SVM model
svm_model = SVR(kernel='linear')
svm_model.fit(X_train, y_train)

# Make predictions on the test data
y_pred_svm = svm_model.predict(X_test)

# Calculate the Mean Squared Error (MSE)
mse_svm = mean_squared_error(y_test, y_pred_svm)
print(f"Mean Squared Error (SVM): {mse_svm}")

# Calculate the Root Mean Squared Error (RMSE)
rmse_svm = np.sqrt(mse_svm)
print(f"Root Mean Squared Error (SVM): {rmse_svm}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np


# Flatten X_train and X_test
X_train_flattened = X_train.reshape(X_train.shape[0], -1)
X_test_flattened = X_test.reshape(X_test.shape[0], -1)

# Normalize the data
scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train_flattened)
X_test_normalized = scaler.transform(X_test_flattened)

# Initialize and train the Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_normalized, y_train)

# Make predictions on the test data
y_pred_rf = rf_model.predict(X_test_normalized)

# Calculate the Mean Squared Error (MSE)
mse_rf = mean_squared_error(y_test, y_pred_rf)
print(f"Mean Squared Error (Random Forest): {mse_rf}")

# Calculate the Root Mean Squared Error (RMSE)
rmse_rf = np.sqrt(mse_rf)
print(f"Root Mean Squared Error (Random Forest): {rmse_rf}")
```
